```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='traffic_density_level', data=df)
plt.title("Traffic Density Distribution")
plt.show()
```

Most traffic observations fall under medium density, indicating moderate congestion is most common.

```python
sns.barplot(x='time_of_day', y='traffic_density_level', data=df)
plt.title("Traffic vs Time of Day")
plt.show()
```

Traffic is higher during evening compared to morning, indicating peak congestion after work hours.

```python
sns.barplot(x='weather_condition', y='traffic_density_level', data=df)
plt.title("Traffic vs Weather Condition")
plt.xticks(rotation=45)
plt.show()
```

Traffic density increases during bad weather conditions like rain or fog.

```python
sns.barplot(x='road_type', y='traffic_density_level', data=df)
plt.title("Traffic vs Road Type")
plt.show()
```

Highways show lower congestion compared to city roads due to better traffic flow.

```python
sns.barplot(x='day_of_week', y='traffic_density_level', data=df)
plt.xticks(rotation=45)
plt.title("Traffic vs Day of Week")
plt.show()
```

Weekdays show higher congestion than weekends due to office traffic.

1. Traffic peaks during evening hours
2. Bad weather increases congestion
3. Weekdays have higher traffic than weekends
