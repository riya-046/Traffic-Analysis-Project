```python
import pandas as pd

df = pd.read_csv('delhi_traffic_features.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['Trip_ID'], inplace=True)
```

```python
df.columns
```

```python
df['time_of_day'] = df['time_of_day'].str.lower().str.strip()
df['day_of_week'] = df['day_of_week'].str.lower().str.strip()
df['weather_condition'] = df['weather_condition'].str.lower().str.strip()
df['road_type'] = df['road_type'].str.lower().str.strip()
df['traffic_density_level'] = df['traffic_density_level'].str.lower().str.strip()
```

```python
mapping = {
    'low': 1,
    'medium': 2,
    'high': 3
}

df['traffic_density_level'] = df['traffic_density_level'].map(mapping)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = [
    'start_area',
    'end_area',
    'time_of_day',
    'day_of_week',
    'weather_condition',
    'road_type'
]

for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

Data Cleaning Steps:

- Removed duplicates
- Dropped Trip_ID column
- Standardized text columns
- Converted traffic levels to numeric
- Encoded categorical variables
